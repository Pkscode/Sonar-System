```python
X=data.iloc[:,0:60].values
```

```python
y=data.iloc[:,60].values
```

```python
# feature Encoding
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
encoder=LabelEncoder()
y=encoder.fit_transform(y)
```

```python
# Splitting into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)  
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# ! pip install keras
# Building an ANN 
# importing KERAS libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
#Initialize the ANN
model=Sequential()

#Build the input and hidden layers with dropout
model.add(Dense(units=32, activation="relu", kernel_initializer="uniform", input_dim=60))
model.add(Dropout(rate=0.1))

#Adding the second hidden layer
model.add(Dense(units=32, activation="relu"))
model.add(Dropout(rate=0.1))

#Adding an Output Layer
model.add(Dense(units=1, activation="sigmoid" ))

# remember kernel_initializer is used only once in first layer
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
#fitting ANN to training set
model.fit(X_train, y_train, batch_size =10, epochs=100)
```

```python
#predicting on test set
y_pred = model.predict(X_test)>0.5
```

```python
y_pred
```

```python
#evaluating using confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
```
